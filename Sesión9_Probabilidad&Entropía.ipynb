# **Actividad 3. Entrop칤a**

<div align="justify">

La entrop칤a de un sistema aislado es m치xima en el equilibrio y desde el punto de vista macrosc칩pico, un sistema en equilibrio se caracteriza por una distribuci칩n (configuraci칩n) de mayor probabilidad. El sistema avanza espont치neamente a trav칠s de varios macroestados con valores crecientes de probabilidad hacia el macroestado con el mayor n칰mero de configuraciones. La mec치nica estad칤stica esta relacionada con la probabilidad de diferentes estados energ칠ticos y la entrop칤a estad칤stica se define como:


$\small S=k_Bln\Omega$

Donde $\small k_B$ es la constante de Boltzman $\small 1,380649칑10^{23} J/K.$,  y  $\small \Omega$ representa el peso estad칤stico (칩 multiplicidad) que puede presentar el sistema. Se define como:


$\small \Omega =\frac{N!}{n_i!(N-n_i)!}$

Donde  ($\small N!$) representa el numero de todas las configuraciones (modos diferentes) de ordenar  ($\small N$) objetos diferentes (tambi칠n conocido como permutaciones de  ($\small N$)),  ($\small n_1!$) representa el numero de configuraciones diferentes (modos diferentes) de modo que produzcan el mismo macroestado para el objeto ($\small n_1$). 

Calculo de la entrop칤a: Efecto del Numero part칤culas:

En la siguiente actividad calcularemos la entrop칤a del proceso redistribuci칩n de las part칤culas de gas mostrado en la figura 1. La rutina que se presenta en la siguiente celda de c칩digo determina el valor de Omega cuando una part칤cula de gas ha migrado del costado izquierdo de la caja al costado derecho. 

a) Fije en el numero total de part칤culas de gas en 8 y corra la simulaci칩n. Realice captura de pantalla de la curva obtenida. Verifique la informaci칩n presentada en la tabla 1. b) Repita el paso anterior cambiando el numero total de part칤culas de gas a 100, 250, 500 y 1000. Realice captura de pantalla de las curvas obtenidas. c) 쮺ual es el macroestado mas probable? d) Compare el valor de multiplicidad para el ensayo de 8 y 1000 part칤culas de gas. e) Realice una discusi칩n sobre las diferencias presentadas por las curvas obtenidas para cada ensayo. 

</div>

```python
n=int(input("Ingrese el numero de part칤culas de gas (1-1000) = "))
k=np.arange(0,n+1,1)
KB=1.380649e-23 #J/K
C=comb(n,k)
s=sum(C)
S=KB*np.log(C)

fig = make_subplots(rows=1, cols=3)

fig.add_trace(go.Scatter(x=k, y=C, name="Multiplicidad"),row=1, col=1)
fig.add_trace(go.Scatter(x=k, y=C*100/s, name="Probabilidad (%)"), row=1, col=2)
fig.add_trace(go.Scatter(x=k, y=S,name="Entrop칤a J/k)"), row=1, col=3)

fig.update_layout(
    xaxis_title="Numero de particulas en costado izquierdo de la caja",
    title=f"Valores de entrop칤a y probabilidad para {N} part칤culas despu칠s que se alcanza el equilibrio",
    width=1200, height=500
)

fig.show()
```
